```python
def plot_rel_change(ATP_data, dATP_data, start = 5, stop = 50):
    rel_change = []
    # max(min(ATP_data[:,0]), min(dATP_data[:,0]))
    interp_range = np.linspace(start, stop, 350)
    ATP_smoothed = np.interp(interp_range, np.hstack((0,ATP_data[:,0])), np.hstack((0,ATP_data[:,2])))
    dATP_smoothed = np.interp(interp_range, np.hstack((0,dATP_data[:,0])), np.hstack((0,dATP_data[:,2])))
    ATP_smoothed_upper = np.interp(interp_range, np.hstack((0,ATP_data[:,0])), np.hstack((0,ATP_data[:,3])))
    dATP_smoothed_upper = np.interp(interp_range, np.hstack((0,dATP_data[:,0])), np.hstack((0,dATP_data[:,3])))
    ATP_smoothed_lower = np.interp(interp_range, np.hstack((0,ATP_data[:,0])), np.hstack((0,ATP_data[:,1])))
    dATP_smoothed_lower = np.interp(interp_range, np.hstack((0,dATP_data[:,0])), np.hstack((0,dATP_data[:,1])))
    
    return interp_range, dATP_smoothed/ATP_smoothed, dATP_smoothed_upper/ATP_smoothed_upper, dATP_smoothed_lower/ATP_smoothed_lower
            
x, change, change_upper, change_lower = plot_rel_change(ATP_data, dATP_data, start = 18, stop = 50)
plt.plot(x,change)
plt.plot(x,change_upper)
plt.plot(x,change_lower)
```

```python
cluster_df = pd.read_csv(file_name, delim_whitespace=True)
```

```python
#cluster_weights = cluster_df.C1.value_counts().values/30000
```

```python
cluster_weights = cluster_df.C1.value_counts()[cluster_df.C1.value_counts().index!=9] / np.sum(cluster_df.C1.value_counts()[cluster_df.C1.value_counts().index!=9])
```

```python
weighted_average_ATP = np.zeros(350)
weighted_lower_ATP = np.zeros(350)
weighted_upper_ATP = np.zeros(350)
for w, avg, ua, la in zip(cluster_weights, average_array_ATP, upper_array_ATP, lower_array_ATP):
    weighted_average_ATP += w * avg
    weighted_lower_ATP += w*la
    weighted_upper_ATP += w*ua
```

```python
weighted_average_dATP = np.zeros(350)
weighted_lower_dATP = np.zeros(350)
weighted_upper_dATP = np.zeros(350)
for w, avg, ua, la in zip(cluster_weights, average_array_dATP, upper_array_dATP, lower_array_dATP):
    weighted_average_dATP += w * avg
    weighted_lower_dATP += w*la
    weighted_upper_dATP += w*ua
```
